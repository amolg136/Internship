```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

# Q1

```python
webpage1 = requests.get('https://en.wikipedia.org/wiki/Main_Page')
soup1 = BeautifulSoup(webpage1.content)
header1 = soup1.find_all('span',class_="mw-headline")
df_header1 = []
for i in soup1.find_all('span',class_="mw-headline"):
    df_header1.append(i.text)
df_header1
```

# Q3

```python
webpage3 = requests.get('https://www.icc-cricket.com/rankings/mens/team-rankings/odi')
soup3 = BeautifulSoup(webpage3.content)
table3 = soup3.find("table", class_="table")
header3 = table3.find_all("th")

list3 = []

for i in header3:
    title3 = i.text
    list3.append(title3)    

df3 = pd.DataFrame(columns = list3)
print(df3)
```

# Q6

```python
webpage6 = requests.get('https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles')
soup6 = BeautifulSoup(webpage6.content)
Paper_Title = []
for i in soup6.find_all("h2", class_="sc-1qrq3sd-1 gRGSUS sc-1nmom32-0 sc-1nmom32-1 btcbYu goSKRg"):
    Paper_Title.append(i.text)
Paper_Title

Authors = []
for i in soup6.find_all("span", class_="sc-1w3fpd7-0 dnCnAO"):
    Authors.append(i.text)
Authors

Published_Date = []
for i in soup6.find_all("span", class_="sc-1thf9ly-2 dvggWt"):
    Published_Date.append(i.text)
Published_Date

df6 = pd.DataFrame({'Paper Title':Paper_Title, 'Authors':Authors, 'Published Date':Published_Date})
df6
```

# Q7

```python
webpage7 = requests.get('https://www.dineout.co.in/delhi-restaurants/buffet-special')
soup7 = BeautifulSoup(webpage7.content)
Restaurant_Name = []
for i in soup7.find_all("div", class_="restnt-info cursor"):
    Restaurant_Name.append(i.text)
Restaurant_Name

Cuisine7 = []
for i in soup7.find_all("span", class_="double-line-ellipsis"):
    Cuisine7.append(i.text)
Cuisine7

Location7 = []
for i in soup7.find_all("div", class_="restnt-loc ellipsis"):
    Location7.append(i.text)
Location7

Ratings7 = []
for i in soup7.find_all("div", class_="restnt-rating-widget pull-right text-right"):
    Ratings7.append(i.text)
Ratings7

Image_URL = []
for i in soup7.find_all("img", class_="no-img"):
    Image_URL.append(i['data-src'])
Image_URL

df6 = pd.DataFrame({'Restaurant Name':Restaurant_Name, 'Cuisine':Cuisine7, 'Location':Location7, 'Ratings':Ratings7, 'Image URL':Image_URL})
df6
```
