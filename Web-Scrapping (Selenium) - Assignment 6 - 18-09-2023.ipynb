# Q1

```python
driver = webdriver.Chrome()
driver.get("https://www.shine.com/")
```

```python
# Once site is opened, close the pop-up & click on "Job title, skills"

role = driver.find_element(By.CLASS_NAME,"form-control")
#role = driver.find_element(By.CLASS_NAME,"searchBarInput")
role.send_keys('Data Analyst')

location = driver.find_element(By.XPATH,"/html/body/div[1]/div[4]/div/div[2]/div[2]/div/form/div/div[1]/ul/li[2]/div/input")
location.send_keys('Bangalore')

search = driver.find_element(By.XPATH,'//button[@type="submit"]')
search.click()
```

```python
job_title = []
job_location = []
company_name = []
exp_required = []

title_tags = driver.find_elements(By.XPATH,'//h2[@itemprop="name"]')
for i in title_tags[0:10]:
    title=i.text
    job_title.append(title)
    
location_tags = driver.find_elements(By.XPATH,'//div[@class=" jobCard_jobCard_lists_item__YxRkV jobCard_locationIcon__zrWt2"]')
for i in location_tags[0:10]:
    location=i.text
    job_location.append(location)
    
company_tags = driver.find_elements(By.XPATH,'//div[@class="jobCard_jobCard_cName__mYnow"]')
for i in company_tags[0:10]:
    company=i.text
    company_name.append(company)

experience_tag = driver.find_elements(By.XPATH,'//div[@class=" jobCard_jobCard_lists_item__YxRkV jobCard_jobIcon__3FB1t"]')
for i in experience_tag[0:10]:
    experience=i.text
    exp_required.append(experience)
    
df = pd.DataFrame({'Job Title':job_title, 'Job Location':job_location, 'Company Name':company_name, 'Exp Required':exp_required})
df
```

# Q8

```python
driver = webdriver.Chrome()
driver.get("https://www.azquotes.com/")
```

```python
input_val = driver.find_element(By.XPATH, '//a[@href="/top_quotes.html"]')
input_val.click()
```

```python
quote = []
author = []
type_of_quotes = []

start = 1
end = 10
for pages in range(start, end):

    quote_tag = driver.find_elements(By.XPATH,'//a[@class="title"]')
    for i in quote_tag[0:1000]:
        title=i.text
        quote.append(title)
    
    author_tag = driver.find_elements(By.XPATH,'//div[@class="author"]')
    for i in author_tag[0:1000]:
        title=i.text
        author.append(title)
    
    type_of_quotes_tag = driver.find_elements(By.XPATH,'//div[@class="tags"]')
    for i in type_of_quotes_tag[0:1000]:
        title=i.text
        type_of_quotes.append(title)
    
    next_button = driver.find_element(By.XPATH, '//li[@class="next"]')
    next_button.click()
    time.sleep(2)
```

```python
print(len(quote), len(author), len(type_of_quotes))
```

```python
df = pd.DataFrame({'Quote':quote, 'Author':author, 'Type Of Quotes':type_of_quotes})
df
```
