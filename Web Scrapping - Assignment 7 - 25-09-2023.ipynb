```python
url = "https://www.amazon.in/"
driver = webdriver.Chrome()
driver.get(url)
```

```python
input_val = driver.find_element(By.XPATH, '//div[@class="nav-search-field "]//input')
input_val.send_keys('guitar')

search_val = driver.find_element(By.XPATH, '//div[@class="nav-search-submit nav-sprite"]/span')
search_val.click()
```

```python
product_url = []

for pages in range(0, 1):
    url = driver.find_elements(By.XPATH, '//a[@class="a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"]')
    for i in url:
        product_url.append(i.get_attribute("href"))
    next_button = driver.find_element(By.XPATH, '//a[@class="s-pagination-item s-pagination-next s-pagination-button s-pagination-separator"]')
    next_button.click()
```

```python
brand_name = []

for url in product_url:
    driver.get(url)
    
    brand_name_tag = driver.find_elements(By.XPATH, '//*[@id="productOverview_feature_div"]/div/table/tbody/tr[1]/td[2]/span')
    for i in brand_name_tag[0:10]:
        title=i.text
        brand_name.append(title)
```

```python
print(len(brand_name))
```

```python
df = pd.DataFrame({"Brand Name":brand_name})
df
```

# Q2

```python
for url in product_url:
    driver.get(url)

    brand_name = []
    
    brand_name_tag = driver.find_elements(By.XPATH, '//*[@id="productOverview_feature_div"]/div/table/tbody/tr[1]/td[2]/span')
    for i in brand_name_tag[0:10]:
        title=i.text
        brand_name.append(title)    
    
    prod_name = []

    prod_name_tag = driver.find_elements(By.XPATH,'//span[@id="productTitle"]')
    for i in prod_name_tag[0:10]:
        title=i.text
        prod_name.append(title)

    prod_price = []

    prod_price_tag = driver.find_elements(By.XPATH,'//span[@class="a-price aok-align-center"]')
    for i in prod_price_tag[0:10]:
        title=i.text
        prod_price.append(title)

    retrn_excng = []

    retrn_excng_tag = driver.find_elements(By.XPATH, '//*[@id="RETURNS_POLICY"]')
    for i in retrn_excng_tag[0:10]:
        title=i.text
        retrn_excng.append(title)

    exp_delivery = []

    exp_delivery_tag = driver.find_elements(By.XPATH, '//*[@id="mir-layout-DELIVERY_BLOCK"]')
    for i in exp_delivery_tag[0:10]:
        title=i.text
        exp_delivery.append(title)

    availability = []

    availability_tag = driver.find_elements(By.XPATH, '//*[@id="availability"]')
    for i in availability_tag[0:10]:
        title=i.text
        availability.append(title)

    prod_url = []

    prod_url_tag = driver.find_elements(By.XPATH,'//span[@class="a-price-whole"]')
    for i in prod_url_tag[0:10]:
        title=i.text
        prod_url.append(title)
```

```python
df1 = pd.DataFrame({'Brand Name':brand_name, 'Name of the Product':prod_name, 'Price':prod_price, 'Return/Exchange':retrn_excng, 'Expected Delivery':exp_delivery,'Availability':availability, 'Product URL':prod_url})
df1
```
