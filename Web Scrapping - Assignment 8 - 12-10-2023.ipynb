### Q1 :-  Scrape the details of most viewed videos on YouTube from Wikipedia.

```python
url = "https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos"
driver = webdriver.Chrome()
driver.get(url)

search_val = driver.find_element(By.XPATH, '//*[@id="toc-Historical_most-viewed_videos"]')
search_val.click()
time.sleep(3)
```

```python
video_name = []
artist_name = []
upload_date = []
views = []

video = driver.find_elements(By.XPATH, '//table[@class="wikitable sortable jquery-tablesorter"][2]/tbody/tr/td[1]/a')
for v in video:
    video_name.append(v.text)
    
artist = driver.find_elements(By.XPATH, '//table[@class="wikitable sortable jquery-tablesorter"][2]/tbody/tr/td[2]')
for a in artist:
    artist_name.append(a.text)
    
upload = driver.find_elements(By.XPATH, '//table[@class="wikitable sortable jquery-tablesorter"][2]/tbody/tr/td[4]')
for u in upload:
    upload_date.append(u.text)
    
view = driver.find_elements(By.XPATH, '//table[@class="wikitable sortable jquery-tablesorter"][2]/tbody/tr/td[3]')
for v in view:
    views.append(v.text)
```

```python
df1 = pd.DataFrame({"Video Name": video_name, "Artist Name": artist_name, "Upload Date": upload_date, "Views": views})
df1
```

### Q8 :-  Details of Datasets from UCI machine learning repositories.

```python
url = "https://archive.ics.uci.edu/"
driver = webdriver.Chrome()
driver.get(url)

sub_menu = driver.find_element(By.XPATH, '//div[@class="flex flex-wrap justify-center gap-5"]/a')
sub_menu.click()
```

```python
pop_up = driver.find_element(By.XPATH, '//button[@class="btn-primary btn-sm btn m-1"]')
pop_up.click()
```

```python
Dataset_Name = []
Data_Type = []
Task = []
Attribute_Type = []
No_of_Instances = []
No_of_Attribute = []
Year = []
```

```python
data_url = []

start = 1
end = 3
for page in range(start, end):
    url = driver.find_elements(By.XPATH, '//div[@class="relative col-span-8 sm:col-span-7"]/h2/a')
    for i in url:
        data_url.append(i.get_attribute("href"))
        
    next_button = driver.find_element(By.XPATH, '//div[@class="btn-group"]/button[2]')
    next_button.click()
```

```python
print(len(data_url))
```

```python
for url in data_url:
    driver.get(url)
    time.sleep(3)
        
    try:
        dataset_name = driver.find_elements(By.XPATH, '//h1[@class="text-3xl font-semibold text-primary-content"]')
        for dn in dataset_name:
            Dataset_Name.append(dn.text)
    except NoSuchElementException:
        Dataset_Name.append("-")
        
    try:
        data_type = driver.find_elements(By.XPATH, '//div[@class="grid grid-cols-8 gap-4 md:grid-cols-12"]/div[2]/p')
        for dt in data_type:
            Data_Type.append(dt.text)
    except NoSuchElementException:
        Data_Type.append("-")
        
    try:
        task = driver.find_elements(By.XPATH, '//div[@class="grid grid-cols-8 gap-4 md:grid-cols-12"]/div[3]/p')
        for t in task:
            Task.append(t.text)
    except NoSuchElementException:
        Task.append("-")
     
    try:
        attribute_type = driver.find_elements(By.XPATH, '//div[@class="grid grid-cols-8 gap-4 md:grid-cols-12"]/div[4]/p')
        for at in attribute_type:
            Attribute_Type.append(at.text)
    except NoSuchElementException:
        Attribute_Type.append("-")        

    try:
        no_of_instances = driver.find_elements(By.XPATH, '//div[@class="grid grid-cols-8 gap-4 md:grid-cols-12"]/div[5]/p')
        for noi in no_of_instances:
            No_of_Instances.append(noi.text)
    except NoSuchElementException:
        No_of_Instances.append("-")       
        
    try:
        no_of_attribute = driver.find_elements(By.XPATH, '//div[@class="grid grid-cols-8 gap-4 md:grid-cols-12"]/div[6]/p')
        for noa in no_of_attribute:
            No_of_Attribute.append(noa.text)
    except NoSuchElementException:
        No_of_Attribute.append("-")
        
    try:
        year = driver.find_elements(By.XPATH, '//div[@class="flex flex-col"]/div/div/h2')
        for y in year:
            Year.append(y.text)
    except NoSuchElementException:
        Year.append("-")
```

```python
 print(len(Dataset_Name), len(Data_Type), len(Task), len(Attribute_Type), len(No_of_Instances), len(No_of_Attribute), len(Year))
```

```python
df8 = pd.DataFrame({"Dataset Name":Dataset_Name, "Data Type":Data_Type, "Task":Task, "Attribute Type":Attribute_Type, "No of Instances":No_of_Instances, "No of Attribute":No_of_Attribute})
df8
```
